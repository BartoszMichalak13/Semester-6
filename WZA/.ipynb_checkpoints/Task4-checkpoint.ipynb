```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define theta values
theta = np.linspace(0, 2*np.pi, 1000)

# Define the polar function r(theta)
r = np.sin(2*theta)

# Convert polar coordinates to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot the curve
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$r(\theta) = \sin(2\theta)$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of r(theta) = sin(2theta)')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
