```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

Task 3.1

```python
z1 = lambda x,y: x**2 + y**2
```

```python
x = np.arange(-10,10)
X,Y = np.meshgrid(x,x)

ax = plt.axes(projection="3d")
ax.set_xlabel("x"); ax.set_ylabel("y"); ax.set_zlabel("z")
surf = ax.plot_surface(X, Y, z1(X,Y), linewidth=0, shade=False)
plt.show()
```
