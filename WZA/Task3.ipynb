```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

Task 3.1

```python
z1 = lambda x,y: x**2 + y**2
```

```python
x = np.arange(-10,10)
X,Y = np.meshgrid(x,x)

ax = plt.axes(projection="3d")
ax.set_xlabel("x"); ax.set_ylabel("y"); ax.set_zlabel("z")
surf = ax.plot_surface(X, Y, z1(X,Y), linewidth=0, shade=False)
plt.show()
```

```python
# Define a grid of points
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
x,y = np.meshgrid(x, y)

# Evaluate the equations xz and yz
z_con1 = np.zeros_like(x)
z_con2 = (x == 0) & (y == 0)

fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")

surf1 = ax.plot_surface(x, y, z_con1, color = 'r', linewidth=0, shade=False)
surf2 = ax.plot_surface(x, y, z_con2, color = 'b', linewidth=0, shade=False)


ax.plot([0,0],[0,0],[0,1], color = 'g', linewidth=3, label='x = 0, y = 0')
plt.show()
```
